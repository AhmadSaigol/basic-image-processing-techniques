```python
import cv2
import numpy as np
import sys
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/test.jpg')
img.shape
height, width=img.shape[:2]
```

```python
#translation
T=np.float32([[1, 0, 100,], [0, 1, 200]])
trans_img=cv2.warpAffine(img, T, (width, height))
cv2_imshow(trans_img)
```

```python
#Rotation
R=cv2.getRotationMatrix2D((width/2, height/2), 30, 1 )
rotate_img=cv2.warpAffine(img, R, (width, height))
cv2_imshow(rotate_img)
```

```python
#flipping
flip_image=cv2.flip(img, 1)
cv2_imshow(flip_image)
```

```python
#Scaling
res = cv2.resize(img, (int(width / 2), int(height / 2)), interpolation = cv2.INTER_AREA)
cv2_imshow(res)
```

```python
#Scaling and rotation
S=cv2.getRotationMatrix2D((width/2, height/2), 30, 0.5 )
scale_img=cv2.warpAffine(img, S, (width, height))
cv2_imshow(scale_img)
```

```python
#Shear
Sh=np.float32([[0.75, 0.5, 0,], [0.5, 0.75, 0]])
Shear_img=cv2.warpAffine(img, Sh, (width, height))
cv2_imshow(Shear_img)
```
